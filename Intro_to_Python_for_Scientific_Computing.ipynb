## Using Matplotlib for plotting useful stuff 

```
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

#defaults
plt.plot(X, C)
plt.plot(X, S)

plt.show()
```

```
plt.figure(figsize=(10, 6), dpi=80)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label = "cosine")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-", label = "sine")

plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.ylim(C.min() * 1.1, C.max() * 1.1)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))


font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }

plt.title('My sin and cos functions', fontdict=font)

plt.legend(loc='upper left')

plt.show()
```

# Plotting Quiver plots 
Very useful in this course

```
x,y = np.meshgrid(np.arange(-2, 2, .2), np.arange(-2, 2, .25))
z = x*np.exp(-x**2 - y**2)
# z is the function of x,y

v, u = np.gradient(z, .2, .2)
fig, ax = plt.subplots(1,2, figsize=(15, 6), dpi=80)

# v is the y component of arrow vectors
# u is the x component of arrow vectors
q = ax[0].quiver(x,y,u,v, edgecolor='k')
q = ax[1].quiver(x,y,u,v, edgecolor='k', width = 0.005)

ax[0].set_title('Thin arrows')
ax[1].set_title('Thicc arrows')


plt.show()
```

# 3D plots and countour plots

```
def my_3d_func(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) 
```

```
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
Z = my_3d_func(X,Y)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)
ax.set_zlim(-2, 2)
ax = plt.gca()
ax.set_title('3D plot')
plt.show()
```

```
plt.contourf(X, Y, Z, cmap= plt.cm.hot) # try other cmaps like 'jet'
plt.contour(X, Y, Z, colors='black')
ax = plt.gca()
ax.set_title('Contour plot')
plt.show()
```

```
# An example on how to use seaborn with matplotlib
import seaborn as sns

X = np.arange(-1, 1.25, 0.25)
Y = np.arange(-1, 1.25, 0.25)
X, Y = np.meshgrid(X, Y)
Z = my_3d_func(X,Y)

fig, ax1 = plt.subplots(1,1)
g=sns.heatmap(Z, ax =ax1)
g.set_xticklabels(np.arange(-1, 1.25, 0.25), rotation = 0, fontsize = 8)
g.set_yticklabels(['a','b','c','d', 'e', 'f', 'g','h','i'], rotation = 0, fontsize = 8)
plt.xlabel("Values on X axis")
plt.ylabel("Values on Y axis")
plt.title('Heatmap of my 3D function')
```

# Working with differential equations in scipy

Solving something like 

dy / dt = f(t, y)

y(t0) = y0

```
import numpy as np
import scipy as sp
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```
'''
Here we have a system of diff equations
dx/dt = ax - b(xy)
dy/dt = -cy + d(xy) 

x,y  = z 
# x and y are in a sense unwrapped from z, 
another way of saying that z is a tuple of x,y
'''

def lotkavolterra(t, z, a, b, c, d):
    x, y = z
    return [a*x - b*x*y, -c*y + d*x*y]
```

We pass in the parameter values a=1.5, b=1, c=3 and d=1 with the args argument.

[0,15] is the time span and x0=10 & y0=5 are initial conditions.

```
sol = solve_ivp(lotkavolterra, [0, 15], [10, 5], method='RK45', args=(1.5, 1, 3, 1), dense_output=True)
```

```
t = np.linspace(0, 15, 300)
z = sol.sol(t)
plt.plot(t, z.T)
plt.xlabel('t')
plt.legend(['x', 'y'], shadow=True)
plt.title('Lotka-Volterra System')

plt.show()
```
